```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("titanic-data.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
Survived_info = []
for i in df['Survived']:
    if i == 0:
        i = 'Not Survived'
        Survived_info.append(i)
    else:
        i = 'Survived'
        Survived_info.append(i)
df['Survived'] = Survived_info
```

```python
df[df['PassengerId'].duplicated()]
```

```python
df = df.dropna(subset=['Embarked'], how='any',axis=0)
```

```python
class_info = []
for i in df['Pclass']:
    if i == 1:
        i = 'Upper'
        class_info.append(i)
    elif i == 2:
        i = 'Middle'
        class_info.append(i)
    elif i == 3:
        i = 'Lower'
        class_info.append(i)
df['Pclass'] = class_info
```

```python
df.head()
```

```python
embarked = []
for i in df['Embarked']:
    if i == 'S':
        i = 'southampton'
        embarked.append(i)
    elif i == 'C':
        i = 'Cherbourg'
        embarked.append(i)
    elif i == 'Q':
        i = 'Queenstown'
        embarked.append(i)
df['Embarked'] = embarked
```

```python
ages = []
for age in df['Age']:
    if age>0 and age<=10:
        age='0-10'
        ages.append(age)
    elif age >10 and age <= 20:
        age = '11-20'
        ages.append(age)
    elif age >20 and age <= 30:
        age = '21-30'
        ages.append(age)
    elif age>30 and age <= 40:
        age = '31-40'
        ages.append(age)
    elif age > 40 and age <= 50:
        age ='41-50'
        ages.append(age)
    elif age >50 and age <=60:
        age = '51-60'
        ages.append(age)
    elif age >60 and age<=70:
        age = '61-70'
        ages.append(age)
    elif age>70 and age<=80:
        age = '71-80'
        ages.append(age)
    else:
        ages.append('None')
df['age'] = ages
```
